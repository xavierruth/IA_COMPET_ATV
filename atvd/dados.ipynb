```python
# Renomeando

colunas = ["PATIENTID", "IDADE", "GENERO", "PAIS", "ANO", "GENETICA", "ARPOLUICAO", "ALCOOL", "FUMO", "OBESIDADE", "TIPOCANCER", "ESTCANCER", "CUSTOUSD", "SOBREVIVENCIA", "ALVOPONTUA"]
gpc.columns = colunas 

# visualizando a tabela com as colunas renomeadas
gpc
```

**Convertendo a variável alvo ("ALVOPONTUA")**

```python
# função para categorizar a gravidade da doença em leve, moderado, severo 

def classificar_gravidade(score):
    if score < 4:
        return 'Leve'
    elif score < 6.5:
        return 'Moderado'
    else:
        return 'Severo'

gpc["GRAVE"] = gpc["ALVOPONTUA"].apply(classificar_gravidade)
```

```python
gpc.head()
```

Analisando através da variável ALVOPONTUA, podemos perceber que a Genética, Fumo, ArPoluição e Alcool indicam que os pacientes com maior risco genético, fumante, excesso de uso de alcool e exposição a poluição do ar tendem a apresentar uma maior tendência a ter uma maior pontuação na gravidade da doença. 

### Gráficos



Perguntas para guiarem a análise: 

1. Há alguma relação entre a alta pontuação na gravidade da doença e a genetica do paciente? (GENETICA X GRAVE)
2. Pacientes fumantes tendem a ter um cancer mais grave? (FUMO X GRAVE)
3. A obesidade influencia na gravidade da doença? (OBSIDADE X GRAVE)
4. A poluição tem um fator colabora no agravamento do cancer? (ARPOLUICAO X GRAVE)
5. Pacientes que ingerem alcool em excesso podem ter um cancer mais grave? (ACOOL X GRAVE)

```python
import matplotlib.pyplot as plt #importando a biblioteca para gerar os gráficos
```

```python
ax = gpc['GRAVE'].hist(figsize=(4, 4))
plt.show()
```

***1.Há alguma relação entre a alta pontuação na gravidade da doença e a genetica do paciente?***

```python

plt.figure(figsize=(8, 4))
sns.boxplot(data=gpc, x="GRAVE", y="GENETICA", hue='GRAVE', palette={'Severo': 'purple', 'Moderado':'blue', 'Leve': 'gray'}, legend=False)


plt.title("Relação entre Gravidade da Doença e Risco Genético", fontsize=14)
plt.xlabel("Gravidade da Doença")
plt.ylabel("Risco Genético")


plt.tight_layout()
plt.show()
```

***2. Pacientes fumantes tendem a ter um cancer mais grave?***

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=gpc, x="GRAVE", y="FUMO", hue='GRAVE', palette={'Severo': 'purple', 'Moderado':'blue', 'Leve': 'gray'}, legend=False)
plt.title("Relação entre Gravidade da Doença e Fumo")
plt.xlabel("Gravidade da Doença")
plt.ylabel("Índice de Fumo")
plt.tight_layout()
plt.show()
```

***3. A obesidade influencia na gravidade da doença?***

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=gpc, x="FUMO", y="GRAVE", hue='GRAVE', estimator=np.mean, palette={'Severo': 'purple', 'Moderado':'blue', 'Leve': 'gray'}, errorbar=None)
plt.title("Média de Obesidade por Gravidade da Doença")
plt.xlabel("Gravidade da Doença")
plt.ylabel("Índice Médio de Obesidade")
plt.tight_layout()
plt.show()
```

***4. A poluição tem um fator colabora no agravamento do cancer?***

```python
pollution_mean = gpc.groupby("GRAVE")["ARPOLUICAO"].mean()
colors = ['gray', 'blue', 'purple']

plt.figure(figsize=(6, 6))
plt.pie(pollution_mean, labels=pollution_mean.index, autopct="%1.1f%%", colors=colors, startangle=140)
plt.title("Proporção Média de Poluição por Gravidade da Doença")
plt.tight_layout()
plt.show()
```

***5. Pacientes que ingerem alcool em excesso podem ter um cancer mais grave?***

```python

alcool_grupo = gpc.groupby("GRAVE")["ALCOOL"].mean()
colors = ['gray', 'blue', 'purple']


labels = alcool_grupo.index
valores = alcool_grupo.values
x = np.arange(len(labels))
width = 0.5

plt.figure(figsize=(8, 4))
plt.bar(x, valores, width=width, label="Álcool", color=colors)
plt.xticks(x, labels)
plt.ylabel("Índice Médio de Álcool")
plt.title("Média de Consumo de Álcool por Gravidade da Doença")
plt.tight_layout()
plt.show()
```
