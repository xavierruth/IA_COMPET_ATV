### Segundo bloco

treinamento oficial, escolhido por ter um melhor retorno na acurácia.

```python
modelos = {
    "Random Forest": RandomForestClassifier(n_estimators=500, class_weight='balanced', random_state=42),

    "Logistic Regresssion": LogisticRegression(max_iter=1000, class_weight='balanced'),

    "SVM": SVC(class_weight='balanced')
}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Armazenar resultados
resultados = {modelo: {"acc": [], "prec": [], "rec": [], "f1": []} for modelo in modelos}

# Treinamento e avaliação 30 vezes
for i in range(30):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    for nome, modelo in modelos.items():
        modelo.fit(X_train_scaled, y_train)
        y_pred = modelo.predict(X_test_scaled)

        resultados[nome]["acc"].append(accuracy_score(y_test, y_pred))
        resultados[nome]["prec"].append(precision_score(y_test, y_pred, average='weighted', zero_division=0))
        resultados[nome]["rec"].append(recall_score(y_test, y_pred, average='weighted'))
        resultados[nome]["f1"].append(f1_score(y_test, y_pred, average='weighted'))
```

```python
print("\n================ RESULTADOS MÉDIOS =================")
for nome in resultados:
    print(f"\nModelo: {nome}")
    for metrica in resultados[nome]:
        media = np.mean(resultados[nome][metrica])
        desvio = np.std(resultados[nome][metrica])
        print(f"{metrica.upper()}: Média = {media:.4f} | Desvio Padrão = {desvio:.4f}")
```

De acordo com a acurácia e o f1 observados, o modelo preditivo escolhido foi o Random Forest, pois indicou um mehor desempenho em relação aos outros dois.

```python
# Supondo que 'tgpc' já esteja carregado com os dados
X = tgpc.drop('GRAVE', axis=1)
y = tgpc['GRAVE']

print("Shape de X:", X.shape)
print("Shape de y:", y.shape)
```

```python
def avaliar_modelo(modelo, X_train, X_test, y_train, y_test):
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)

    acuracia = accuracy_score(y_test, y_pred)
    precisao = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    return [acuracia, precisao, recall, f1]
```

```python
def rodar_avaliacoes(X, y, n_iteracoes=30):
    resultados_modelo1 = []
    resultados_modelo2 = []
    resultados_modelo3 = []

    for _ in range(n_iteracoes):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

        modelo1 = RandomForestClassifier(n_estimators=500, class_weight='balanced', random_state=42)
        modelo2 = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
        modelo3 = SVC(kernel='linear', class_weight='balanced', random_state=42)

        resultados_modelo1.append(avaliar_modelo(modelo1, X_train, X_test, y_train, y_test))
        resultados_modelo2.append(avaliar_modelo(modelo2, X_train, X_test, y_train, y_test))
        resultados_modelo3.append(avaliar_modelo(modelo3, X_train, X_test, y_train, y_test))

    return (
        np.array(resultados_modelo1),
        np.array(resultados_modelo2),
        np.array(resultados_modelo3)
    )
```

```python
def mostrar_resultados(nome_modelo, resultados):
    metricas = ["Acurácia", "Precisão", "Recall", "F1-score"]
    print(f"\nResultados para {nome_modelo}:")
    for i, metrica in enumerate(metricas):
        media = np.mean(resultados[:, i])
        desvio = np.std(resultados[:, i])
        print(f"{metrica}: média = {media:.4f}, desvio padrão = {desvio:.4f}")
```

```python
resultados1, resultados2, resultados3 = rodar_avaliacoes(X, y, n_iteracoes=30)

mostrar_resultados("Random Forest", resultados1)
mostrar_resultados("Logistic Regression", resultados2)
mostrar_resultados("SVM (Linear)", resultados3)
```

De acordo com a acurácia e o f1 observados, o modelo preditivo escolhido foi o Random Forest, pois indicou um mehor desempenho em relação aos outros dois.
